```python
import numpy as np
import pandas as pd

input_dir = 'data/'
train = pd.read_csv(input_dir + 'train.csv',encoding='utf_8')
test = pd.read_csv(input_dir + 'test.csv')
resources = pd.read_csv(input_dir + 'resources.csv')
submission = pd.read_csv(input_dir + 'sample_submission.csv')

resources['total_price'] = resources.quantity * resources.price
resources.head()
train.head()
```

```python
mean_total_price = pd.DataFrame(resources.groupby('id').total_price.mean()) 
mean_total_price.head()
```

```python
#Add id as column for merging
mean_total_price['id'] = mean_total_price.index
train["essay"] = train["project_essay_1"].notnull() + train["project_essay_2"].notnull()+ train["project_essay_3"].notnull() + train["project_essay_4"].notnull()
train = pd.merge(train, mean_total_price, on='id')
X_train = train.total_price.values.reshape(-1, 1) #reshape because of one column
y_train = train.project_is_approved.values

train.head()
```

```python
import spacy
import pandas as pd
import itertools as it

nlp = spacy.load('en')
```

```python
a = nlp(train['project_essay_1'][0])
print a
```

```python
for num, sentence in enumerate(a.sents):
    print 'Sentence {}:'.format(num + 1)
    print sentence
    print ''
```

```python
for num, entity in enumerate(a.ents):
    print 'Entity {}:'.format(num + 1), entity, '-', entity.label_
    print ''
```

```python
token_text = [token.orth_ for token in a]
token_pos = [token.pos_ for token in a]

pd.DataFrame(zip(token_text, token_pos),
             columns=['token_text', 'part_of_speech'])
```

```python
token_lemma = [token.lemma_ for token in a]
token_shape = [token.shape_ for token in a]

pd.DataFrame(zip(token_text, token_lemma, token_shape),
             columns=['token_text', 'token_lemma', 'token_shape'])
```

```python
token_entity_type = [token.ent_type_ for token in a]
token_entity_iob = [token.ent_iob_ for token in a]

pd.DataFrame(zip(token_text, token_entity_type, token_entity_iob),
             columns=['token_text', 'entity_type', 'inside_outside_begin'])
```

```python
token_attributes = [(token.orth_,
                     token.prob,
                     token.is_stop,
                     token.is_punct,
                     token.is_space,
                     token.like_num,
                     token.is_oov)
                    for token in a]

df = pd.DataFrame(token_attributes,
                  columns=['text',
                           'log_probability',
                           'stop?',
                           'punctuation?',
                           'whitespace?',
                           'number?',
                           'out of vocab.?'])

df.loc[:, 'stop?':'out of vocab.?'] = (df.loc[:, 'stop?':'out of vocab.?']
                                       .applymap(lambda x: u'Yes' if x else u''))

df
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
n_folds = 10
kf = KFold(n_splits=n_folds)
Average_AUC = 0
for (train_index, test_index) in kf.split(X_train):
    X, X_val = X_train[train_index], X_train[test_index]
    y, y_val = y_train[train_index], y_train[test_index]
    clf = LogisticRegression()
    clf.fit(X, y)
    pred = clf.predict_proba(X_val)[:,1]
    AUC = roc_auc_score(y_val, pred)
    Average_AUC += AUC
    print("AUC: {}".format(AUC))
Average_AUC = Average_AUC / n_folds
print("Average_AUC: {}".format(Average_AUC))
```

```python
test = pd.merge(test, mean_total_price, on='id')
```

```python
X_test = test.total_price.values.reshape(-1, 1)
pred_test = clf.predict_proba(X_test)[:,1]
```
